```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset("tips")
iris= sns.load_dataset("iris")
```

```python
#Bar Plot- Used For Categorical Data & Is Shows The Count Of Occurances
sns.barplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
#Histogram is a graph that shows the frequency of numerical data.
sns.histplot(tips ['total_bill'])
```

```python
#To see the line of Probabilities Add kde=True
sns.histplot(tips ['total_bill'],kde=True)
```
